```python
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize

##Jika punk belum ada##
# nltk.download('punk')

vartext = "Saya Sedang Berpergian ke Pasar membeli ikan, karena saya suka banget sama ikan\n"
print(vartext)
```

```python
from nltk.probability import FreqDist

# find frequency distribution
fdist = FreqDist(vartext)

# print 10 most common
fdist.most_common(10)
```

```python
import matplotlib.pyplot as plt
fdist.plot(10)
```
